```python
def extractCsv(path,sepCsv,liste,name_liste):
    
    """ Warning : file only contains _ or camelCase no point in name file
        Onyl for csv simple no big file"""
    import os.path

    #import d'OS 
    #listDir = lecture du chemin 
    file_to_extract=os.listdir(path)
    #On itère sur le dossier pour exfilter l'ensemble des fichiers dans ce dossier
    count_liste=0 # permettra de retrouver le DF dans la liste
    try:
        for i in file_to_extract:
            df=pd.DataFrame()
            #On split le nom du fichier (qui ne doit pas contenir de point hors extension, et on prend la première sortie) 
            a=i.split('.')[0]
            try:
                #A chaque itération de boucle on ajoute à la liste le résultat de la lecture du fichier 
                chunk=pd.read_csv(f"{path}{i}",sep=sepCsv,chunksize=500000)
                for i in chunk:
                    df=pd.concat([df,i])
                
                liste.append(df)
                print(f'{a}={name_liste}[{count_liste}]')
                count_liste+=1
            except:
                #Si une erreur est détectée
                print(f"{i} n'est pas un fichier")
           
    except:
        pass
    return liste
```

```python
def ScaleUpdate(df,colDep,pounds,colFin=''):
    try:
        # si une seule colonne
        if colFin=='':
            #si la colonne est un int 
            if type(colDep)==int:
                df.iloc[:,colDep]=df.iloc[:,colDep]*pounds
           #si la colonne est un string
            elif type(colDep)==str:
                df.loc[:,colDep]=df.loc[:,colDep]*pounds

        else : #si plusieurs colonnes 
        #si les colonnes sont des int
            if type(colDep)==int and type(colFin)==int:
                df.iloc[:,colDep:colFin]=df.iloc[:,colDep:colFin]*pounds
        #qi les colonnes sont des strings
            elif type(colDep)==str and type(colFin)==str:
                df.loc[:,colDep:colFin]=df.loc[:,colDep:colFin]*pounds
    except:
        print('Erreur paramètre')
    return df
```
